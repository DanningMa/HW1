```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3, 4])
y1 = np.array([1, 2, 3, 4, 5])
x2 = np.array([0, 1, 2, 3, 4])
y2 = np.array([5, 6, 7, 8, 9])
x3 = np.array([0, 1, 2, 3, 4])
y3 = np.array([9, 10, 11, 12, 13])

plt.plot(x1, y1)
plt.plot(x2, y2)
plt.plot(x3, y3)
plt.xlabel("X Lable")
plt.ylabel("Y Lable")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.array([0, 1, 2, 3])
y = np.array([4, 6, 6, 5])
plt.subplot(2, 2, 1)
plt.plot(x,y,marker = 'D',color = 'r')


x = np.array([0, 1, 2, 3])
y = np.array([5, 6, 6, 4])
plt.subplot(2, 2, 2)
plt.plot(x,y,marker = 'D',color = 'r')

x = np.array([0, 1, 2, 3])
y = np.array([6, 4, 3, 2])
plt.subplot(2, 2, 3)
plt.plot(x,y,marker = '*',color = 'r')

x = np.array([0, 1, 2, 3])
y = np.array([3, 4, 5, 7])
plt.subplot(2, 2, 4)
plt.plot(x,y,marker = '*',color = 'r')

plt.show()
```
