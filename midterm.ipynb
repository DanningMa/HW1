```python
# Question 1 part 1: Find the frequency of each distinct element in the list using a dictionary

list = [2,4,6,8,4,5,2,1,9,0,4,6,7,4,3,2,1,9,10,3,7,9,6,0,1,3,5,6,7,8,9,10,2,3,6,8,9,10,6,7,4,3]
mydict1 = {}
for i in list:
    if i in mydict1:
        mydict1[i] += 1
    else:
        mydict1[i] = 1

mydict2 = {}
for name in sorted(mydict1.keys()):
    mydict2[name] = mydict1[name]

print(mydict2)
```

```python
# Question 1 part 2: Plot the frequency results, e.g. using bar chart

import pandas as pd
import numpy as np
import json 
dict = [2,4,6,8,4,5,2,1,9,0,4,6,7,4,3,2,1,9,10,3,7,9,6,0,1,3,5,6,7,8,9,10,2,3,6,8,9,10,6,7,4,3]
data = {"number": dict}
df = pd.DataFrame(data)
df.groupby("number").size()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.countplot(df["number"],label="count")
plt.show()
```

```python
# Question 1 part 3: Store this dictionary results into json file. 

jasofile = json.dumps(mydict2)
print(jasofile)
```

```python
# Question 2: import the data of historical transactions of Amazon 

import pandas as pd
import matplotlib.pyplot as plt
dt = pd.read_csv('Downloads/All-Amazon-order2.csv')
dt.head()
```

```python
# Question 2: The shape (dimen) of historical transactions of Amazon shows 16 orders from Aug 30 2019 to Mar 22 2021. 

dt.shape
```

```python
# Question 2: Replace all NaN with 0

dt = dt.fillna(0)
dt.head()
```

```python
# Question 2: Delete sign $ in Total Charged.

dt["Total Charged"] = dt["Total Charged"].str.replace('$','').astype(float)
dt.head()
```

```python
# Question 2: Some basic information of Total Charged from Aug 30 2019 to Mar 22 2021.

a = dt["Total Charged"].sum()
b = dt["Total Charged"].mean()
c = dt["Total Charged"].median()
d = dt["Total Charged"].max()
e = dt["Total Charged"].min()
print ('sum of all Total Charged is', a)
print ('mean of all Total Charged is', b)
print ('median of all Total Charged is', c)
print ('max of all Total Charged is', d)
print ('min of all Total Charged is', e)
```

```python
# Question 2: Delete sign $ in Tax Charged.
dt["Tax Charged"] = dt["Tax Charged"].str.replace('$','').astype(float)
dt.head()
```

```python
# Question 2: The sum of Tax charged from Aug 30 2019 to Mar 22 2021.
dt["Tax Charged"].sum()
```

```python
# Question 2: Calculate the tax rate by tax/ total charge
dt["Tax Charged"].sum() / dt["Total Charged"].sum()
# tax rate is about 5.5% which is near the tax in DC 6% 
```

```python
# Question 2: Change the format of Order data
dt['Order Date'] = pd.to_datetime(dt['Order Date'])
dt.head()
```

```python
# Question 2: The plot with x-aixs: order data, y-axis: total charge of each order 
dt.plot.bar(x='Order Date', y='Total Charged', rot=90)
```

```python
# Question 2: Combine the total charge in each day
daily_orders = dt.groupby('Order Date').sum()["Total Charged"]
daily_orders.head()
```

```python
# Question 2: The plot with x-aixs: date of consume, y-axis: total charge in each day
daily_orders.plot.bar(figsize=(15,10), color = 'green')
```
